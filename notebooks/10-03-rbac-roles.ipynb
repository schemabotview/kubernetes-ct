## RBAC — Roles & ClusterRoles

**RBAC** is Kubernetes' permission system — the AUTHZ answer from notebook 08: *can this subject do this verb on this resource?* Two role objects, identical in shape, different in scope:

- **`Role`** — namespaced. Permissions on resources in **one namespace**.
- **`ClusterRole`** — cluster-scoped. Covers cluster-scoped resources (nodes, PVs, namespaces), non-resource URLs (`/healthz`), *or* resources across **every** namespace.

```yaml
kind: Role
metadata: { name: pod-reader, namespace: team-a }
rules:
  - apiGroups: [""]              # "" = core group (Pod, Service, ConfigMap)
    resources: ["pods", "pods/log"]
    verbs: ["get", "list", "watch"]
  - apiGroups: ["apps"]
    resources: ["deployments"]
    verbs: ["get", "list"]
```

Each rule has three parts:

- **`apiGroups`** — `""` (core), `apps` (Deployments), `batch` (Jobs), `networking.k8s.io`, `rbac.authorization.k8s.io`.
- **`resources`** — lowercase plural; sub-resources with `/`: `pods/log`, `pods/exec`, `deployments/scale`.
- **`verbs`** — `get`, `list`, `watch`, `create`, `update`, `patch`, `delete`, `deletecollection`, plus specials `*`, `impersonate`, `bind`/`escalate`.

Common combos: read-only = `get,list,watch`; full = add `create,update,patch,delete`; `kubectl exec/logs` = `get` on `pods/exec`,`pods/log`.

**The "no deny" rule:** RBAC is **purely additive** — you can't write "deny X." Permissions accumulate from every binding naming you; to remove one, delete the binding. On our map these are the **Role** and **ClusterRole** chips in the RBAC box — permission sets that do nothing until *bound* (next section).
