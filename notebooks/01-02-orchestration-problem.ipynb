## The orchestration problem — what Docker alone doesn't solve

You finished the `docker` course, so you can already build images, run containers, mount volumes, wire networks, and bring up a small stack with `docker compose`. So what's missing?

The honest answer: as soon as you have *more than one machine* and *more than one application*, you hit a cluster of problems Docker alone wasn't built to solve.

- **Placement.** Five nodes, twenty containers — which goes where? You want CPU-hungry workloads spread out and small ones packed tight. By hand that's a full-time job. Kubernetes' **scheduler** does it automatically.
- **Self-healing.** A container crashes at 3 a.m. With Docker alone it stays crashed until someone notices. Kubernetes replaces it in seconds — and if a *node* dies, redistributes its containers to the survivors.
- **Scaling.** Two replicas → traffic doubles → you need ten. With `docker run` that's SSH-ing into machines. Kubernetes changes one number — `replicas: 10` — and converges.
- **Rollouts & rollbacks.** Shipping v2 shouldn't mean "stop all v1, then start all v2." Kubernetes does **rolling updates** — start a v2, wait until healthy, stop a v1, repeat. One command rolls back.
- **Service discovery.** `api-7f9c` and `api-aa31` are both your API. The `web` pod shouldn't need to know which exist right now. A stable **Service** name like `api` plus cluster DNS handles it.
- **Configuration & secrets.** Twenty services need the same DB URL; one needs a TLS cert. Kubernetes has first-class **ConfigMaps** and **Secrets** pods reference declaratively.

Docker still does the actual container work. **Kubernetes is what you put on top** — on top of a container runtime, really (notebook 08) — once you have more containers than you can supervise personally.

This whole list maps onto the **Cluster** box on our platform map: the control plane is the machinery that solves placement, healing, scaling, and rollout, while the workers run the pods it schedules.
