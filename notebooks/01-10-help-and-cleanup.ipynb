## Getting help & cleaning up

You won't memorise every flag of every command or every field of every object. You don't need to — Kubernetes ships its own reference.

- **`kubectl --help`** — the top-level command list.
- **`kubectl <verb> --help`** — flags for one verb (`kubectl run --help`).
- **`kubectl explain <resource>`** — the fields of an object kind, offline. Pair it with a dotted path to drill in: `kubectl explain pod.spec.containers`. This *is* the API reference on disk.
- **`kubectl api-resources`** — every kind the cluster knows, with short name, API group, and whether it's namespaced.

For deeper docs: **kubernetes.io/docs** is canonical (its "Concepts" and "Tasks" sections are excellent), and the **CKA is open-book** against exactly those pages. The flow when stuck: `kubectl explain` for "what fields," `kubectl <verb> --help` for "what flags," then the docs site for tutorials.

### Cleaning up

A few hours in and you'll have stray pods, leftover deployments, and orphaned configmaps. Two ways to clean:

- **Targeted** — `kubectl delete pod <name>`, or `kubectl delete -f pod.yaml` to remove whatever a file describes.
- **Bulk** — `kubectl delete <kind> --all` clears every object of that kind in the namespace. **Take care** — it really does delete.

And the whole cluster, when you're done for the day:

```bash
kind delete cluster --name learn
```

The cluster comes down in seconds. That throwaway-ness is *the point* of `kind` for learning — break it, delete it, recreate it, no consequences. On our map you've now seen the whole **kubectl** column of verbs; the rest of the course zooms the camera into each subsystem it drives.
