## The pod — Kubernetes' smallest unit

Here's the first surprise coming from Docker. Kubernetes' smallest deployable thing is **not a container**. It's a **pod**.

A **pod is one or more containers that share a network and a storage scope, scheduled together onto the same node, and treated as a single unit.**

For 95% of pods there is exactly one container inside, so "pod" and "container" *feel* interchangeable. The abstraction matters because of the other 5%.

- **Shared network.** All containers in a pod share one IP and port space. They reach each other on `localhost`. From outside, the pod has one IP.
- **Shared storage scope.** Volumes are declared on the pod and mounted into whichever containers need them — so a sidecar can write to a directory the main container reads.
- **Same node, same lifecycle.** A pod is scheduled atomically onto one node. Its containers start, live, and die there together.

**Why bother?** The pod exists for *helper containers*:

- **Sidecars** — a companion that adds a capability (log shipping, a service-mesh proxy).
- **Init containers** — run-once containers that prepare the pod before the main container starts (fetch config, wait for a DB).
- **Adapter / ambassador** — older names for sidecars that translate or proxy traffic.

One container → the pod is a thin wrapper. A main process plus one or two helpers that *must* share a machine and filesystem → the pod is the right unit. In this notebook and the next, every pod runs exactly one container to keep the noise down. Multi-container patterns arrive in notebook 02. On our map, this is the **Pods** box on the worker node — the thing everything else ultimately serves.
