## Networking, Ingress & network policies

Notebook 04 gave you Services — L4 routing to pods. This module goes deeper: the **flat IP-per-pod model** underneath, the **CNI** that makes it real, **Ingress** for L7 HTTP routing, and **NetworkPolicy** — the cluster's built-in firewall.

The through-line is that Kubernetes networking is a small set of axioms plus pluggable enforcement:

- **The model** — four hard requirements every cluster satisfies: every pod a routable IP, pod-to-pod with no NAT, node-to-pod, and one consistent IP per pod.
- **Pod CIDRs & the CNI** — where addresses come from, and what the plugin does in the milliseconds a pod is born.
- **Three communication paths** — pod↔pod, pod↔Service, pod↔external — each routed differently. Almost every "why unreachable?" is "which path, which layer?"
- **Ingress** — one public entry point fanning out to many Services by host/path, with TLS. The resource vs. the controller that obeys it. Plus its successor, the **Gateway API**.
- **NetworkPolicy** — the default-open cluster made default-deny, additively, enforced by the CNI.

On our map this whole module is the **Networking** box in the Resources band — **Service**, **Ingress**, **NetworkPolicy**, **CoreDNS** — plus the **CRI/CNI** on the worker node that plumbs it. Where notebook 04 was "how do I reach a workload," this is "how does the packet actually travel, and how do I lock it down."
