## ConfigMaps, Secrets & environment

Notebook 03 ran an image with the config baked in. But you want *one* image that runs in staging with `LOG_LEVEL=debug` and in prod with `LOG_LEVEL=warn`, points at a swappable backend URL, and mounts the same TLS cert into three services. Baking config into images breaks that — you'd ship a different, differently-tested image per environment. The fix is a **first-class object that holds configuration and gets referenced from pod templates.**

Two objects, near-identical schemas:

- **ConfigMap** — non-sensitive key/value config (strings or whole files).
- **Secret** — the same shape, plus a *security posture* (base64 storage, `tmpfs` mounts, optional etcd encryption, tighter RBAC).

> Master ConfigMap, and Secret is the same thing plus handling.

Both are consumed **three ways** — as individual env vars, as a bulk env import, or as files in a volume — and the choice controls **whether config live-updates** without a restart. We'll cover both objects, immutability, image-pull secrets, and the **downward API** (a pod reading facts about *itself*).

On our map this module lives in the **Config** box of the Resources band — **ConfigMap**, **Secret**, and **env / mount** — plus one reach up to the **Registry** for image-pull secrets. The through-line: **decouple config from the image** so the artifact you tested is the artifact you ship.
