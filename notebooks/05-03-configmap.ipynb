## ConfigMap — the object

A **ConfigMap** is a namespace-scoped key/value store of configuration that pods reference by name. Values can be short strings or whole files.

```yaml
apiVersion: v1
kind: ConfigMap
metadata: { name: app-config }
data:
  LOG_LEVEL: "info"
  GREETING: "hello from staging"
  app.properties: |
    server.port=8080
    cache.ttl=300s
```

Two sections under the top level:

- **`data`** — string values (UTF-8, no binary). Where nearly every ConfigMap lives.
- **`binaryData`** — base64-encoded blobs (a keystore, an image). Less common.

Key idea: **a ConfigMap is just data.** It has **no `spec` and no `status`** — nothing for a controller to reconcile. The cluster does *nothing* with it until a pod *references* it. That's the difference from every object so far: a Deployment drives pods; a ConfigMap just sits there being read.

### The 1 MiB limit

A ConfigMap (and a Secret) is capped at **1 MiB** total — an `etcd` limit, since every object must fit in one `etcd` value. For a 10 MB lookup table, use object storage, a database, or a sidecar — not a ConfigMap.

You create them declaratively (the YAML above) or imperatively — `kubectl create configmap app-config --from-literal=LOG_LEVEL=info --from-file=app.properties`. On our map this is the **ConfigMap** chip in the Config box: inert data, waiting to be projected into a Pod by one of the three consumption patterns we cover next.
