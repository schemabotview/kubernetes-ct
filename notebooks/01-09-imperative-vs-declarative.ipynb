## Imperative vs declarative

`kubectl run hello --image=nginx:alpine` is an **imperative** command — you say *what to do right now*. It's fast for learning, fine for ad-hoc tasks, and accepted on the exam under time pressure. But it's not how real systems are managed.

The grown-up way is **declarative**: write a YAML file describing the object you want, then ask Kubernetes to make reality match.

```yaml
# pod.yaml
apiVersion: v1
kind: Pod
metadata:
  name: hello
  labels:
    app: hello
spec:
  containers:
    - name: nginx
      image: nginx:alpine
      ports:
        - containerPort: 80
```

```bash
kubectl apply -f pod.yaml
```

Why declarative wins for anything non-throwaway:

- **Version-controllable.** YAML lives in Git — diff, review, roll back.
- **Idempotent.** Apply the same file ten times → one pod. `kubectl run` errors on the second try.
- **Composable.** A directory of YAML describes a whole environment: `kubectl apply -f ./manifests/`.
- **Reusable.** Helm and Kustomize template these files across environments.

The bridge between the two — and a CKA time-saver — is to *generate* YAML from an imperative command:

```bash
kubectl run hello --image=nginx:alpine --dry-run=client -o yaml > pod.yaml
```

`--dry-run=client` means "build the object but don't send it"; `-o yaml` prints it. You get a working manifest in one shot to edit and `apply`. Most CKA candidates write their YAML this way — faster and harder to typo.

On our map, `kubectl apply` is the **apply** verb writing your desired object into the **api-server**, which stores it in etcd — the same declarative door notebook 02 walks through field by field.
