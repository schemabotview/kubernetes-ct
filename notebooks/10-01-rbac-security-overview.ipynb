## RBAC, security, troubleshooting & CKA prep

The final module. Nine modules built the platform; this one **locks it down** and **teaches you to fix it** — the two halves the CKA weighs most.

**Security** — from identity to policy:

- **Identity** — Users (people, not objects) vs **ServiceAccounts** (pods, real objects). RBAC applies to both.
- **RBAC** — `Role`/`ClusterRole` (permissions) bound to subjects by `RoleBinding`/`ClusterRoleBinding`. Purely additive, no "deny." `kubectl auth can-i` verifies.
- **Hardening** — the pod **security context** (non-root, drop caps, read-only fs), **Pod Security Admission** enforcing a level per namespace, **etcd encryption at rest**, and image/supply-chain basics.

**Troubleshooting** — ~30% of the exam: a broken cluster, you fix it. Know the failure modes and the first command for each — **application** (`Pending`, `CrashLoopBackOff`), **control plane & node** (api-server down, `NotReady`), and **networking** (Service unreachable).

Then **CKA exam strategy** — time budget, aliases, the imperative `--dry-run=client -o yaml` muscle memory, and a drill list.

On our map this module frames the **RBAC** and **Security** boxes of the Operate band, then tours the *whole* map as troubleshooting playbooks. The final section steps back to the full platform — a fitting close, because by now you can name every box on it.
