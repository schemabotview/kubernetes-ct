## The cluster at 30,000 feet

A Kubernetes cluster is split into two kinds of machines: the **control plane** and the **worker nodes**. Each runs a specific set of programs, they talk over the network, and together they form one logical "cluster."

```
       Control plane                    Worker node
+------------------------+     +-------------------------+
|  etcd  <—  api-server  | <—> |  kubelet                |
|            ^     ^     |HTTPS|     |                    |
|  scheduler-+     +-----|     |  container runtime      |
|  controller-manager    |     |     |  (containerd)      |
+------------------------+     |     v                   |
                               |  your containers (pods) |
                               |  kube-proxy             |
                               +-------------------------+
```

Don't memorise the boxes yet. Take away the shape:

- **`etcd`** — a distributed key-value store. The cluster's source of truth. Every object you create ends up as a row here.
- **`kube-apiserver`** — the front door. Everything that reads or changes cluster state goes through it. **Nothing talks to `etcd` directly except the API server.**
- **`kube-scheduler`** — decides which node a new pod lands on.
- **`kube-controller-manager`** — runs the background loops that watch the world and make it match desired state.
- **`kubelet`** — on every worker node. Gets told "run these pods" and tells the local runtime to do it.
- **`kube-proxy`** — also on every node. The network plumbing that makes Services work.
- **Container runtime** — what actually executes containers, usually `containerd` (from the docker course). The **CRI** (container runtime interface) is how the kubelet talks to it.

We open each of these properly in notebook 08. For now: **the control plane decides; the nodes execute.** On our map, that's the whole **Cluster** column — control plane on the left, worker node on the right.
