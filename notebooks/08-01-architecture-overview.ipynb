## Cluster architecture & kubeadm

Seven modules in, you've *used* the control plane. Now we open it up — the machinery behind every object you've created. This is the densest chapter in the course and the CKA tests it heavily.

The architecture, with every piece named:

```
Control plane                         Worker nodes
 etcd ←→ kube-apiserver ← kubectl      kubelet ─CRI→ runtime → your pods
           ↑ watch                     kube-proxy    CNI (network)
 scheduler   controller-manager
             cloud-controller-manager
```

Four rules of thumb to memorise:

- **Only the API server talks to `etcd`.** Everything else talks to the API server.
- **The API server is stateless** — horizontally scalable; state lives in `etcd`.
- **Controllers don't poll, they watch** — a long-lived HTTPS stream delivers events.
- **Nodes don't trust each other** — everything authenticates to the API server, the single source of trust.

The module walks each box — **etcd, api-server, scheduler & controllers, node components (kubelet/kube-proxy/runtime), the CRI/CNI/CSI plug-in points** — then the operational skills: **static pods**, **kubeadm** install, the **cluster PKI**, **etcd backup/restore** (the CKA's flag-bearer), and **cluster upgrades**. On our map this whole module *is* the **Control Plane** and **Node runtime** boxes — the two halves of the Cluster column that everything else has been riding on.
