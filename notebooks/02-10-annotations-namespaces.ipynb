## Annotations & namespaces

**Annotations** are the other half of metadata. Labels are for *selection*; annotations are for *everything else you attach to an object*. Same key/value shape, at `metadata.annotations`, but with no length limit and **no use by selectors**.

```yaml
metadata:
  annotations:
    owner: platform-team@example.com
    description: "Front-door HTTP service for the public site"
    git-commit: abc123def
```

Uses: human notes, build metadata (git commit, image digest), and **tool configuration** — Ingress controllers, meshes, and operators read annotations to toggle features without changing the object's schema (`nginx.ingress.kubernetes.io/rewrite-target`). `kubectl apply` even stores its last-applied state in one. Rule of thumb: **find it later → label; describe it or feed a tool → annotation.**

### Namespaces — soft tenancy

A **namespace** scopes object names: `default/web` and `staging/web` are distinct. Most kinds are namespaced; a few (Nodes, PersistentVolumes, StorageClasses, ClusterRoles) are cluster-scoped. A namespace gives you a **name scope**, an **RBAC target**, a **quota/limit target**, and a **DNS suffix** (`<svc>.<ns>.svc.cluster.local`).

It does **not** give you **network isolation** (every pod can reach every pod until a NetworkPolicy says otherwise — notebook 09) or **strong tenancy** (soft boundaries for teams within one trusted org, not hostile multi-tenancy).

Built-ins: **`default`** (your stuff), **`kube-system`** (CoreDNS, kube-proxy, controllers), **`kube-public`** (world-readable), **`kube-node-lease`** (node heartbeats). Switch per-command with `-n <ns>`, or make it stick:

```bash
kubectl config set-context --current --namespace=team-a
```

On our map, namespaces are a slicing of the whole **Cluster** — the same objects, partitioned — which is why this section frames the Cluster box rather than a single node.
