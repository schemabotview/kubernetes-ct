## What Kubernetes is — and the shape of the whole platform

If you ask ten people what Kubernetes is, you get ten answers. The cleanest one-line definition:

> **Kubernetes runs containers across a fleet of machines for you — it restarts them when they die, replaces them when a machine dies, and gives them a stable way to find each other.**

That sentence does a lot of work. Pick it apart:

- **A fleet of machines.** Kubernetes assumes more than one server — physical, virtual, on-prem, or cloud. Each is a **node**. A small group runs Kubernetes itself (the **control plane**); the rest run your containers (**worker nodes**).
- **Runs containers for you.** You don't say "start this container on that server." You say "I want three copies running *somewhere*" and Kubernetes picks where. That decision is **scheduling**.
- **Restarts them when they die.** A container crashes → Kubernetes starts a new one. A *machine* dies → Kubernetes replaces its containers on the survivors. You declare desired state; Kubernetes maintains it.
- **Stable way to find each other.** Pod IPs change constantly. A **Service** gives a stable name and IP that always points at whichever pods are alive.

The name is Greek for "helmsman." **k8s** = *k*, eight letters, *s*. "Kube" means the same thing.

**Why it matters.** In 2026, containers in production mostly means *Kubernetes* in production — the managed big three (**EKS**, **GKE**, **AKS**), on-prem (OpenShift, Rancher, `kubeadm`), laptops (`kind`, `minikube`), and CI/CD. Kubernetes is the default, and the **CKA** (Certified Kubernetes Administrator) is the credential this course takes you toward.

**The shape of the whole platform** — the one map every module in this course zooms into:

```
  Client (kubectl)  →  Cluster  →  Registry
                     ┌───────────────────────┐
                     │ Control Plane          Worker Node │
                     │ apiserver·etcd·sched   kubelet·proxy·CRI │
                     │ controllers            → Pods           │
                     └───────────────────────┘
   Workloads · Resources (storage/net/config) · Operate (sched/RBAC/security)
```

Everything points at the **api-server** at the center: kubectl writes to it, controllers watch it, the kubelet executes its decisions, and every workload, resource, and policy resolves onto the **Pods** it governs. Hold that picture — the rest of the course is a tour of it.
