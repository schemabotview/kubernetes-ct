## The core pod lifecycle

A pod moves through a small set of **phases**:

```
  kubectl run / apply
         |
         v
     Pending  ---- scheduled, image pulling, init containers
         |
         v
     Running  ---- at least one container up
         |
    +----+----+
    v         v
 Succeeded   Failed        (exit 0 vs non-zero)

     Unknown  ---- node lost contact with the API server
```

The everyday commands for living with a pod:

- **`kubectl get pods`** — list pods. `-w` to watch live, `-o wide` for node and IP, `-A` for all namespaces.
- **`kubectl describe pod <name>`** — events, image, IP, conditions, the last restart reason. The single most useful "why is my pod sad" command.
- **`kubectl logs <name>`** — stdout/stderr. `-f` to follow, `--previous` for the *previous* instance after a crash.
- **`kubectl exec -it <name> -- <cmd>`** — run a command inside the container. The `--` separates `kubectl` flags from your command.
- **`kubectl port-forward <name> <local>:<pod>`** — tunnel a local port to a pod port. Handy before a Service exists.
- **`kubectl delete pod <name>`** — delete it. A **bare** pod stays gone; a pod **owned by a controller** gets a replacement within seconds — because the controller is reconciling desired state.

That last contrast is the whole lesson: a bare pod is mortal; a controller-managed pod is *resurrected*. Most commands also accept label selectors — `-l app=web` — to act on every matching pod (labels arrive in notebook 02). On our map, all of this orbits the **Pods** box: the verbs on the left (`logs`, `exec`, `describe`) are how you observe it; its phases are what the kubelet reports back up to the api-server.
