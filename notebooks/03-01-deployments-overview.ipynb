## Deployments, ReplicaSets & rollouts

Notebook 02 created pods directly. That's fine for learning and hopeless for production — a bare pod dies with its node and can't be updated without downtime. This module introduces the objects that fix both: the **controllers** that own pods.

The chain to hold in your head:

```
Deployment  →owns→  ReplicaSet  →owns→  Pods
 (you write)        (per version)      (the running copies)
```

- **ReplicaSet** — keeps exactly `N` pods matching a selector alive. Great at *count*, useless at *change*.
- **Deployment** — owns ReplicaSets, one per version. It choreographs the switch between them: a **rolling update** with history and one-command **rollback**.

Everything here is the reconcile loop from notebook 02, made concrete: you edit a `spec`, and a controller drives reality to match — creating pods when count is low, spinning up a new ReplicaSet when the template changes, scaling the old one down as the new one proves healthy.

We'll build up ReplicaSet → Deployment → scaling → rolling updates → rollback → strategies, then tour the **rest of the workload family** (DaemonSet, StatefulSet, Job, CronJob). On our map this is the **Workload kinds** box and the **Rollout** box in the Workloads band — the machinery that turns "I want three healthy copies of v2" into running Pods. For the CKA, the three-layer ownership (Deployment → ReplicaSet → Pod) is one of the most-tested pictures — memorise it.
