## Volumes — the basics

A pod declares a list of **volumes** under `spec.volumes`; each container mounts any of them under `volumeMounts`. The link is **by name**.

```yaml
spec:
  containers:
    - name: app
      image: busybox
      volumeMounts:
        - name: scratch          # references the volume below
          mountPath: /workspace
  volumes:
    - name: scratch              # the name
      emptyDir: {}               # the type
```

Two things to internalise:

- **Declared on the pod, mounted into the container.** That's how containers *share* a volume — the sidecar pattern from notebook 02: both declare a `volumeMount` referencing the same `volumes[].name`.
- **A "volume type" is one key** under the volume entry: `emptyDir: {}`, `hostPath:`, `configMap:`, `secret:`, `persistentVolumeClaim:`. Each entry uses **exactly one** type key.

The mount supports a few knobs: **`readOnly: true`**, **`subPath`** (mount only a sub-path — the "doesn't live-update" gotcha from notebook 05), and **`mountPropagation`** (advanced CSI/host sharing, rarely by hand).

The types split into two big groups: **ephemeral** (live with the pod) and **persistent** (outlive it, backed by a PV). The rest of the module walks both. On our map, a volume is the wire between a **Pod** and whatever backs it — an ephemeral scratch space, or the durable **Storage** box below.
