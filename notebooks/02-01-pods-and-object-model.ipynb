## Pods, labels & the object model

Notebook 01 ran a pod and toured the platform. This module opens up the **object** you actually create — the Pod — and the **model** every Kubernetes object shares.

Three ideas carry the whole notebook:

- **Every object has the same five keys.** `apiVersion`, `kind`, `metadata`, `spec`, `status`. Learn them once and no manifest ever feels foreign — a Pod, a Service, a Deployment, a Secret all read the same way.
- **`spec` is what you want; `status` is what is.** You write `spec`; controllers write `status`; the gap between them is the work the cluster is always doing. That reconcile loop *is* Kubernetes.
- **Labels turn a flat list into a queryable set.** Once you have more than a handful of pods, you stop naming them and start *selecting* them: "everything with `app=web`." Services, Deployments, NetworkPolicies — all of them find pods this way.

We'll build the Pod up field by field — its manifest, its phases and probes, its init and sidecar containers — then step out to **labels, selectors, annotations, and namespaces**: the metadata layer that lets you organise thousands of objects without losing your mind.

On our platform map, this whole module lives in the **Workloads** band — the **Pod** box and its parts (`containers`, `initContainers`, `sidecar`, `probes`) — with a detour up to the **api-server**, where the object model itself is enforced. The Pod is the atom everything else in the course is built from; this is where we learn its anatomy.
