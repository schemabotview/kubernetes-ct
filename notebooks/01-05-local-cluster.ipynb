## Getting a local cluster

You only learn Kubernetes by running pods, and for that you need a cluster. Four good options, roughly by simplicity.

**Option 1 — `kind` (Kubernetes IN Docker).** One binary; each Kubernetes "node" runs as a Docker container. Boots in under a minute, easy to reset.

```bash
brew install kind                    # macOS
kind create cluster --name learn     # single-node cluster
kind delete cluster --name learn     # wipe it
```

Fastest start, multi-node works, heavily used in CI. Nodes are containers, not VMs — fine for learning.

**Option 2 — `minikube`.** Single-node cluster in a VM or container, with bundled add-ons (`minikube addons enable ingress`). Very forgiving; slower to boot.

**Option 3 — Docker Desktop's bundled Kubernetes.** A checkbox in settings — "Enable Kubernetes." Zero new installs if you already run Docker Desktop; not always the latest version.

**Option 4 — `kubeadm` on VMs.** The "real" install — `kubeadm init` on one node, `kubeadm join` on the rest. Matches production exactly and is what the CKA expects. We do this end-to-end in notebook 08. Heaviest setup; not where you start.

**For notebooks 01–07, use `kind`** — the lightest path to a working cluster.

**You also need `kubectl`**, the client that talks to the cluster:

```bash
brew install kubectl
```

Verify the cluster is healthy with three commands:

- `kubectl version` — client *and* server versions means `kubectl` can reach the API server.
- `kubectl cluster-info` — the API server URL and add-on endpoints.
- `kubectl get nodes` — lists nodes; with `kind` you'll see one named `learn-control-plane`.

On our map, spinning up a cluster is standing up that whole **Cluster** column locally — one node that is *both* control plane and worker.
