## Scheduling — resources, affinity, taints

When a controller creates a pod, its `spec.nodeName` is empty and it sits `Pending`. The **`kube-scheduler`** decides which node it lands on — and this module is about the **inputs that bias that decision.**

The scheduler runs one loop: **filter** nodes that *could* run the pod → **score** the survivors → **bind** the pod to the best. Everything here tunes one of those steps:

- **Requests & limits** — how much you reserve (filter) and cap (runtime), plus the **QoS** class that decides eviction order.
- **`nodeSelector` & node affinity** — "I want *this kind of node*" (pod-side, node labels).
- **Pod affinity / anti-affinity & topology spread** — "near / away from *these other pods*."
- **Taints & tolerations** — "*stay away* from this node unless you opted in" (node-side).
- **PriorityClass & preemption** — "I'm important enough to *kick others out*."
- **Governance** — `LimitRange` and `ResourceQuota` for multi-team clusters.

The scheduler **decides**; the kubelet **runs**. It's a controller like any other — read state, act, write `spec.nodeName`. On our map this whole module is the **Scheduling** box in the Operate band, all feeding the **kube-scheduler** in the control plane. Master these inputs and you control *where* every Pod runs.
