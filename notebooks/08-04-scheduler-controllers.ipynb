## `kube-scheduler` and `kube-controller-manager`

Two control-plane workhorses you've already met.

**`kube-scheduler`** (recap from notebook 07) — watches `Pending` pods, filters and scores nodes, binds the pod to the winner. Pluggable via profiles. Runs as a static pod; talks to the API server, never to nodes.

**`kube-controller-manager`** — one binary hosting ~**30 controllers**, each the same observe-compare-act loop. The ones that matter:

| Controller | Does |
|---|---|
| **Node** | Marks nodes `NotReady` when heartbeats stop; manages `NoExecute` taints. |
| **Deployment / ReplicaSet** | Rollouts; pods to match the count (notebook 03). |
| **DaemonSet / StatefulSet / Job / CronJob** | Manage their pods. |
| **Endpoints / EndpointSlice** | Ready pod IPs behind each Service (notebook 04). |
| **ServiceAccount** | Default SA + token per namespace. |
| **Namespace** | Cascades deletes on `Terminating`. |
| **PV / PVC binder** | Matches claims to volumes (notebook 06). |
| **HPA** | Adjusts replicas by metrics. |

Bundling is just packaging — each controller runs its own loop, and they **elect a leader** (a lease in `kube-system`) so only one instance is active even when the binary runs HA.

**`cloud-controller-manager`** — the cloud-specific bits, split out so non-cloud clusters don't need them: the cloud-side Node controller, the Route controller (pod-CIDR routing), and the Service controller (provisions LoadBalancers — notebook 04). On `kind` it isn't running at all. On our map these are the **kube-scheduler** and **controller-manager** components — the "decide and reconcile" half of the control plane.
