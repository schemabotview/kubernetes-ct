## CKA exam strategy & where to go next

The CKA is **2 hours, ~15–20 hands-on tasks**, scored by automated scripts in a browser terminal. You fix real clusters.

**Time budget.** Read the whole question first — the `kubectl config use-context <foo>` at the top is **mandatory** (skip it and you fix the wrong cluster). **Skip-and-return** on hard tasks (fixed points each; don't sink 30 min into a 4-pointer). Save 10 minutes to re-check.

**Aliases** worth setting up front:

```bash
alias k=kubectl
export do='--dry-run=client -o yaml'   # k run app --image=nginx $do
export now='--grace-period=0 --force'
source <(kubectl completion bash); complete -o default -F __start_kubectl k
```

**The single most important muscle memory** — generate, edit, apply:

```bash
kubectl run x --image=nginx --dry-run=client -o yaml > x.yaml
vim x.yaml && kubectl apply -f x.yaml
```

Same with `create deployment/service/configmap/secret/role/rolebinding/cronjob`. Drill these to **under 5 minutes each, no docs**: Deployment+Service, probes, ConfigMap as env+volume, a Role bound to an SA, a default-deny NetworkPolicy, **etcd snapshot + restore**, rolling update + undo, drain/cordon/uncordon, and diagnosing `Pending`/`CrashLoopBackOff`/`ImagePullBackOff`. Common gotchas: **wrong namespace**, **wrong context**, **selector/template mismatch**, `delete pod` doesn't respect PDBs but `drain` does.

**You're at the end.** From "never run a pod" to a 10-module map of Kubernetes — primitives to cluster operation. Everything rests on four ideas you now own: **declarative state, controllers that reconcile, label selectors, and one API server.** Next: **CKAD** (app-side), **CKS** (security, needs CKA first), the **Gateway API**, service meshes, multi-cluster. New tools speak the same language. On our map, you can now name **every box** — the whole platform, end to end. Good luck with the exam.
